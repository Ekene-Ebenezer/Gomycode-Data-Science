# Summary of the Dataset Profiling

---

Based on the result from the Ydata Profiling;
>There are no missing or null values among the 107,833 observations in the dataset.

>All columns, except for the arrival time column, are of the object data type.

>Ydata has flagged a concern regarding the number of zeros in the dataset, specifically 38,168 (35.4%) zeros. These zeros are attributed to the arrival date column, suggesting that Tunisia flights have achieved a remarkable 35.4% zero-delay rate, aligning with the standard practice of minimizing delays for airline customers.

>The arrival delay column exhibits a maximum delay of 3,451 minutes, indicating the presence of outliers. This outlier is also evident in the mean delay arrival time of 48 minutes, with a median delay of 14 minutes. This suggests that while the majority of flights experience relatively low delays, there are instances of significantly longer delays, contributing to the high mean delay.
