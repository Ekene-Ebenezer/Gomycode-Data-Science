# - Data Exploration with Ydata

```python
# Import the Ydata_profiling library that have been installed
import ydata_profiling as ydp

# Imports profile for creating html or notebook from ydata_profiling results
import profile

#Generates the report
from ydata_profiling import ProfileReport
```

```python
#Creates a Ydata profile with the Title "Education Attainment Profile"

Education_DataProfile = ProfileReport(EDUCATION_Dataset, title = " Education Attainment Profile", minimal = True)
```

```python
#Generates the profile

Education_DataProfile
```

```python
#using grouo by function. This is not part of the code
EDUCATION_Dataset.groupby(["country", "year"]).mean()
```

```python
#Converts the profile created into a html file
Education_DataProfile.to_file("Edu_DataProfile")
```
